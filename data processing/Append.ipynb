```python
import os
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import mlab
import matplotlib.cm as cm
import pickle
from scipy.misc import imresize
import SignalFunctions
import math
import gc
import sys
import threading
import pandas as pd
```

```python
img_rows=80
imagesize=80
fs=250
num=int(2.5*fs)
date="BumblebeeKitchen_08_10_2018"
basefolder='/homes/fl305/research/code/datafile/kitchen1/all'+'/'+date+"/"
if os.path.exists(basefolder)==False:
    os.makedirs(basefolder)
```

```python
def appendf1(datafolder,outfolder,bratio,eratio):
    xfinal1=np.array([]).reshape(0,imagesize, imagesize,2)
    xfinal2=np.array([]).reshape(0,img_rows,2)
    yfinal1=np.array([]).reshape(0, 1)
    file_num=0
    fs= [f for f in os.listdir(datafolder) if os.path.isfile(os.path.join(datafolder,f))]
    for f in fs[int(len(fs)*bratio):int(len(fs)*eratio)]:
        file_num=file_num+1
        print(f)
        save_path=os.path.join(datafolder, f)
        batch=pd.read_pickle(save_path)
        x1=batch['Spectrogram'].values
        x2=batch['scatterize'].values
        y1=batch['behaviour'].values
        batch_size=y1.shape[0]
        for n in range(0, batch_size):
            xn1=x1[n]
            xn2=x2[n]
            xout1=xn1.reshape(1,imagesize, imagesize, 2)
            xout2=xn2.reshape(1,img_rows,2)
            yout1=np.array([[y1[n]]])
           
            xfinal1=np.append(xfinal1,xout1,axis=0)
            xfinal2=np.append(xfinal2,xout2,axis=0)
            yfinal1=np.append(yfinal1,yout1,axis=0)
           
        gc.collect()  
    print(yfinal1.shape)
    if os.path.exists(outfolder)==False:
        os.makedirs(outfolder)
    output = open(outfolder+str(int(bratio*100))+'_xfinal1.p', 'wb')
    pickle.dump(xfinal1, output)
    output = open(outfolder+str(int(bratio*100))+'_xfinal2.p', 'wb')
    pickle.dump(xfinal2, output)
    output = open(outfolder+str(int(bratio*100))+'_yfinal1.p', 'wb')
    pickle.dump(yfinal1, output)
```

```python
def appendf(datafolder,outfolder,bratio,eratio):
    xfinal=np.array([]).reshape(0,imagesize, imagesize,2)
    signal_final=np.array([]).reshape(0,4,num)
    yfinal1=np.array([]).reshape(0, 1)
    file_num=0
    fs= [f for f in os.listdir(datafolder) if os.path.isfile(os.path.join(datafolder,f))]
    for f in fs[int(len(fs)*bratio):int(len(fs)*eratio)]:
        file_num=file_num+1
        print(f)
        save_path=os.path.join(datafolder, f)
        batch=pd.read_pickle(save_path)
        x1=batch['Spectrogram'].values
        signals=batch['signals'].values
        y1=batch['behaviour'].values
        batch_size=y1.shape[0]
        for n in range(0, batch_size):
            xn1=x1[n]
            xout1=xn1.reshape(1,imagesize, imagesize, 2)
            xfinal=np.append(xfinal,xout1,axis=0)
            signalsn=np.reshape(signals[n],(1,4,num))
            yout1=np.array([[y1[n]]])
            signal_final=np.append(signal_final,signalsn,axis=0)
            yfinal1=np.append(yfinal1,yout1,axis=0)
        gc.collect()  
    print(yfinal1.shape)
    if os.path.exists(outfolder)==False:
        os.makedirs(outfolder)
    output = open(outfolder+str(int(bratio*100))+'_Signalall.p', 'wb')
    pickle.dump(signal_final, output)
    
    output = open(outfolder+str(int(bratio*100))+'_specall.p', 'wb')
    pickle.dump(xfinal, output)
    
    output = open(outfolder+str(int(bratio*100))+'_yfinall.p', 'wb')
    pickle.dump(yfinal1, output)
```

```python
outfolder=basefolder+"cluster/"
t1=threading.Thread(target=appendf,args=(basefolder,outfolder,0,0.25,))
t2=threading.Thread(target=appendf,args=(basefolder,outfolder,0.25,0.5,))
t3=threading.Thread(target=appendf,args=(basefolder,outfolder,0.5,0.75,))
t4=threading.Thread(target=appendf,args=(basefolder,outfolder,0.75,1,))
t1.start()
t2.start()
t3.start()
t4.start()
```

```python
xs0=["0_specall.p","25_specall.p","50_specall.p","75_specall.p"]
xs1=["0_Signalall.p","25_Signalall.p","50_Signalall.p","75_Signalall.p"]
ys=["0_yfinall.p","25_yfinall.p","50_yfinall.p","75_yfinall.p"]
xsall=np.array([]).reshape(0,imagesize, imagesize,2)
for i in xs0:
    batch=pd.read_pickle(outfolder+i)
    xsall=np.append(xsall,batch,axis=0)
print(xsall.shape)
output = open(outfolder+'xsall.p', 'wb')
pickle.dump(xsall, output)

xall=np.array([]).reshape(0, 4,num)
for i in xs1:
    batch=pd.read_pickle(outfolder+i)
    xall=np.append(xall,batch,axis=0)
print(xall.shape)
output = open(outfolder+'xalls.p', 'wb')
pickle.dump(xall, output)


yall=np.array([]).reshape(0, 1)
for i in ys:
    batch=pd.read_pickle(outfolder+i)
    yall=np.append(yall,batch,axis=0)
print(yall.shape)
output = open(outfolder+'yall.p', 'wb')
pickle.dump(yall, output)
output.close()
```
